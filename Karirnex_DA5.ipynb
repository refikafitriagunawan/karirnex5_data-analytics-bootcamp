```python
df['sales_date_std'] = pd.to_datetime(
    df['sales_date_std'],
    unit = 'D',
    origin = '1899-12-30'
)
```

```python
df.head()
```

```python
df['revenue_tier'] = np.where(
    df['total_sales'] < 500000,
    'Low',
    np.where(
        df['total_sales'] > 2000000,
        'Mid',
        'High'
    )
)
```

```python
df.head()
```

```python
df_profit = df.groupby(
    ['category', 'kota']
)['total_sales'].mean().reset_index()

df_profit.sort_values(
    by='total_sales',
    ascending=False
)
```

```python
from IPython.display import display

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

pivot = pd.pivot_table(
    df[df['status'] == 'cancelled'],
    values='total_sales',
    index='kota',
    columns='category',
    aggfunc='sum',
    fill_value=0
)
pivot['Total Cancelled'] = pivot.sum(axis=1)
pivot = pivot.sort_values(
    by='Total Cancelled',
    ascending=False
)

display(pivot)
```

```python
revenue_kota = (
    df.groupby('kota')['total_sales']
    .sum()
    .sort_values(ascending=False)
    .head(8)
)

# Membuat bar chart
plt.figure(figsize=(10,5))
bars = plt.bar(revenue_kota.index, revenue_kota.values)

# Judul dan label
plt.title('Top 8 Revenue per Kota')
plt.xlabel('Kota')
plt.ylabel('Revenue')

# Rotasi nama kota
plt.xticks(rotation=45)

# Menambahkan label di atas bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f'{int(height):,}',
        ha='center',
        va='bottom'
    )

plt.tight_layout()
plt.show()
```

```python
# Filter data tahun 2025
df_2025 = df[df['sales_date_std'].dt.year == 2025]
# Membuat total sales per bulan
monthly_sales = (
    df_2025
    .groupby(df_2025['sales_date_std'].dt.month)['total_sales']
    .sum()
)
# Cari bulan dengan sales tertinggi
max_month = monthly_sales.idxmax()
max_value = monthly_sales.max()

# Membuat line chart
plt.figure(figsize=(10,5))
plt.plot(
    monthly_sales.index,
    monthly_sales.values,
    marker='o'
)

# Highlight titik tertinggi
plt.scatter(
    max_month,
    max_value,
    s=150
)

# Label titik tertinggi
plt.text(
    max_month,
    max_value,
    f'Tertinggi: {int(max_value):,}',
    ha='center',
    va='bottom'
)

# Pengaturan chart
plt.title('Trend Total Sales Perbulan Tahun 2025')
plt.xlabel('Bulan')
plt.ylabel('Total Sales')
plt.xticks(range(1,13))

plt.grid(True)

plt.show()
```
